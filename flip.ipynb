```python
product_name=[]
prices=[]


```

```python
r = requests.get(url, headers=headers)

soup = BeautifulSoup(r.text, "lxml")

names=soup.find_all("div",class_="KzDlHZ")
print(names)

for i in names:
    name=i.text
    product_name.append(name)
print(product_name)
```

```python
pric=soup.find_all("div",class_="Nx9bqj _4b5DiR")
print(names)

for i in pric:
    price=i.text
    prices.append(price)
print(prices)
```

```python
description=[]
d=soup.find_all("ul",class_="G4BRas")
print(d)
for i in d:
    p=i.text
    description.append(p)
print(description)
print(len(description))
```

```python
reviews=[]
r=soup.find_all("div",class_="XQDdHH")

for i in r:
    pr=i.text
    reviews.append(pr)
print(reviews)
reviews=reviews[:24]
print(len(reviews))

```

```python
print(len(reviews))
```

```python
df=pd.DataFrame({
    "product name":product_name,
    "price":prices,
    "description":description,
    "review":reviews
})
print(df)
```
